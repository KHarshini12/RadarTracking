```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("radar_tracking_data.csv")
n = len(df) # n = 360 
early_df = df.iloc[:int(n * 0.33)] # 360 *0.33 = 118
middle_df = df.iloc[int(n * 0.33):int(n * 0.67)]
late_df = df.iloc[int(n * 0.67):]
def compute_performance(df_phase):
    return {
        "mae_distance_a": np.mean(np.abs(df_phase["error_a_m"])),
        "mae_distance_b": np.mean(np.abs(df_phase["error_b_m"])),
        "mae_speed_a": np.mean(np.abs(df_phase["speed_a_ms"] - df_phase["true_speed_ms"])),
        "mae_speed_b": np.mean(np.abs(df_phase["speed_b_ms"] - df_phase["true_speed_ms"])),
    }
results = {
    "Early Phase": compute_performance(early_df),
    "Middle Phase": compute_performance(middle_df),
    "Late Phase": compute_performance(late_df),
}
results_df = pd.DataFrame(results)
print("Tracking Performance (Mean Absolute Error):")
print(results_df)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
results_df.loc[["mae_distance_a", "mae_distance_b"]].T.plot(kind='bar', ax=axes[0])
axes[0].set_title("Mean Absolute Distance Error")
axes[0].set_ylabel("Error (m)")
axes[0].set_xticklabels(results_df.columns, rotation=0)
results_df.loc[["mae_speed_a", "mae_speed_b"]].T.plot(kind='bar', ax=axes[1])
axes[1].set_title("Mean Absolute Speed Error")
axes[1].set_ylabel("Error (m/s)")
axes[1].set_xticklabels(results_df.columns, rotation=0)


plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("radar_tracking_data.csv")
df["total_error_a"] = np.abs(df["error_a_m"]) + np.abs(df["speed_a_ms"] - df["true_speed_ms"])
df["total_error_b"] = np.abs(df["error_b_m"]) + np.abs(df["speed_b_ms"] - df["true_speed_ms"])
max_error_a_idx = df["total_error_a"].idxmax()
max_error_b_idx = df["total_error_b"].idxmax()
def classify_phase(index, total_length):
    if index < total_length * 0.33:
        return "Early"
    elif index < total_length * 0.67:
        return "Middle"
    else:
        return "Late"
total_length = len(df)
worst_a = df.loc[max_error_a_idx, ["time_s", "total_error_a"]].copy()
worst_a["phase"] = classify_phase(max_error_a_idx, total_length)
worst_b = df.loc[max_error_b_idx, ["time_s", "total_error_b"]].copy()
worst_b["phase"] = classify_phase(max_error_b_idx, total_length)
print("Worst Error - System A:", worst_a)
print("Worst Error - System B:", worst_b)
plt.plot(df["time_s"], df["total_error_a"], label="System A Total Error", color="blue")
plt.plot(df["time_s"], df["total_error_b"], label="System B Total Error", color="green")
plt.scatter(worst_a["time_s"], worst_a["total_error_a"], color='red', s=100,
            label=f"Max Error A ({worst_a['phase']})", edgecolors='black')
plt.scatter(worst_b["time_s"], worst_b["total_error_b"], color='orange', s=100,
            label=f"Max Error B ({worst_b['phase']})", edgecolors='black')
plt.title("Tracking Performance: Total Error Over Time")
plt.xlabel("Time (s)")
plt.ylabel("Total Error (|distance error| + |speed error|)")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("radar_tracking_data.csv")  # Make sure the file is in the same directory

# Calculate absolute error for System A range
df['abs_error_range_a'] = df['error_a_m'].abs()

# Define window sizes
window_sizes = [10, 20, 30]

# Prepare dictionary to store rolling statistics
rolling_stats = {}

# Calculate rolling mean, std dev, and RMSE
for window in window_sizes:
    rolling_mean = df['abs_error_range_a'].rolling(window=window).mean()
    rolling_std = df['abs_error_range_a'].rolling(window=window).std()
    rolling_rmse = np.sqrt(df['abs_error_range_a'].rolling(window=window)
                           .apply(lambda x: np.mean(x**2), raw=True))
    
    rolling_stats[window] = {
        'mean': rolling_mean,
        'std': rolling_std,
        'rmse': rolling_rmse
    }


plt.figure(figsize=(14, 6))

for window in window_sizes:
    plt.plot(df['time_s'], rolling_stats[window]['mean'], label=f'Mean (Window={window})')
    plt.plot(df['time_s'], rolling_stats[window]['std'], linestyle='--', label=f'Std Dev (Window={window})')
    plt.plot(df['time_s'], rolling_stats[window]['rmse'], linestyle='-.', label=f'RMSE (Window={window})')

plt.title('Rolling Mean, Std Dev, and RMSE - System A Range Error')
plt.xlabel('Time (s)')
plt.ylabel('Error Value')





plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(42)
n_points = 300
true_range = np.linspace(100, 200, n_points)
true_velocity = np.linspace(30, 60, n_points)

estimated_range_A = true_range + np.random.normal(0, 5, n_points)
estimated_velocity_A = true_velocity + np.random.normal(0, 3, n_points)
abs_error_range_A = np.abs(true_range - estimated_range_A)
abs_error_velocity_A = np.abs(true_velocity - estimated_velocity_A)
df_A = pd.DataFrame({
    'abs_error_range': abs_error_range_A,
    'abs_error_velocity': abs_error_velocity_A
})
window_sizes = [10, 20, 30]
rolling_stats = {}
for w in window_sizes:
    rolling_mean = df_A['abs_error_range'].rolling(window=w).mean()
    rolling_std = df_A['abs_error_range'].rolling(window=w).std()
    t_score = stats.t.ppf(0.975, df=w-1)
    
    ci_lower = rolling_mean - t_score * (rolling_std / np.sqrt(w))
    ci_upper = rolling_mean + t_score * (rolling_std / np.sqrt(w))
    
    rolling_stats[w] = {
        'mean': rolling_mean,
        'std': rolling_std,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper
    }
fig, axs = plt.subplots(3, 1, figsize=(14, 14), sharex=True)
x = np.arange(n_points)
for idx, w in enumerate(window_sizes):
    mean_arr = rolling_stats[w]['mean'].to_numpy()
    ci_lower_arr = rolling_stats[w]['ci_lower'].to_numpy()
    ci_upper_arr = rolling_stats[w]['ci_upper'].to_numpy()
    
    axs[idx].plot(x, df_A['abs_error_range'], label='Absolute Error (Range)', color='gray', alpha=0.5)
    axs[idx].plot(x, mean_arr, label=f'Rolling Mean (Window={w})', color='blue')
    axs[idx].fill_between(x, ci_lower_arr, ci_upper_arr, color='blue', alpha=0.2, label='95% Confidence Interval')
    axs[idx].set_title(f'Rolling Mean and 95% CI - Window Size = {w}')
    axs[idx].set_ylabel('Absolute Error')
    axs[idx].legend()
axs[2].set_xlabel('Time Step')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_1samp
import pandas as pd
df['error_distance'] = df['system_a_m'] - df['true_distance_m']
df['error_speed'] = df['speed_a_ms'] - df['true_speed_ms']
t_stat_dist, p_value_dist = ttest_1samp(df['error_distance'], 0)
t_stat_speed, p_value_speed = ttest_1samp(df['error_speed'], 0)
print("Range (Distance) Bias Test:")
print("t-statistic:", t_stat_dist)
print("p-value:", p_value_dist)
if p_value_dist < 0.05:
    print("Result: System A is biased in distance measurement.")
else:
    print("Result: System A is unbiased in distance measurement.")
print("\nVelocity (Speed) Bias Test:")
print("t-statistic:", t_stat_speed)
print("p-value:", p_value_speed)
if p_value_speed < 0.05:
    print("Result: System A is biased in speed measurement.")
else:
    print("Result: System A is unbiased in speed measurement.")
```

```python
from scipy.stats import ttest_1samp
import pandas as pd
df['error_distance_b'] = df['system_b_m'] - df['true_distance_m']
df['error_speed_b'] = df['speed_b_ms'] - df['true_speed_ms']
t_stat_dist_b, p_value_dist_b = ttest_1samp(df['error_distance_b'], 0)
t_stat_speed_b, p_value_speed_b = ttest_1samp(df['error_speed_b'], 0)
print("System B - Range (Distance) Bias Test:")
print("t-statistic:", t_stat_dist_b)
print("p-value:", p_value_dist_b)
if p_value_dist_b < 0.05:
    print("Result: System B is biased in distance measurement.")
else:
    print("Result: System B is unbiased in distance measurement.")
print("\nSystem B - Velocity (Speed) Bias Test:")
print("t-statistic:", t_stat_speed_b)
print("p-value:", p_value_speed_b)

if p_value_speed_b < 0.05:
    print("Result: System B is biased in speed measurement.")
else:
    print("Result: System B is unbiased in speed measurement.")
```

```python
from scipy.stats import ttest_rel
import pandas as pd
import numpy as np

df['abs_error_a'] = np.abs(df['error_a_m'])
df['abs_error_b'] = np.abs(df['error_b_m'])

t_stat, p_value = ttest_rel(df['abs_error_a'], df['abs_error_b'])

print("Comparison of Absolute Errors Between System A and System B")
print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Result: Significant difference between System A and B absolute errors.")
else:
    print("Result: No significant difference between System A and B absolute errors.")
```
